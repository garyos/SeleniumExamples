## Running a google search through chrome

```python
# -*- coding: utf-8 -*-
"""
Created on Sat Jul 27 11:05:37 2019

@author: g_osm
"""

from selenium import webdriver
from selenium.webdriver.common.keys import Keys

# create a new chrome session (path added manually to save restart)
driver = webdriver.Chrome(r'C:\Users\g_osm\chromedriver_win32/chromedriver')


#geckodriver needed to be added to path in order for this to work
#driver = webdriver.Firefox()
driver.implicitly_wait(30)
driver.maximize_window()

# Navigate to the application home page
driver.get("http://www.google.com")

# get the search textbox
search_field = driver.find_element_by_name('q')

search_field.clear()

# enter search keyword and submit
search_field.send_keys("Selenium WebDriver Interview questions")
search_field.submit()

# get the list of elements which are displayed after the search
# currently on result page using find_elements_by_class_name method
lists= driver.find_elements_by_class_name("LC20lb")

# get the number of elements found
print ("Found " + str(len(lists)) + " searches:")

# iterate through each element and print the text that is
# name of the search

i=0
for listitem in lists:
   print (listitem.get_attribute("innerHTML"))
   i=i+1
   if(i>10):
      break

# close the browser window
driver.quit()
```

## With Pagination and Browser Navigation

```python
#geckodriver needed to be added to path in order for this to work
driver = webdriver.Firefox()
driver.implicitly_wait(30)
driver.maximize_window()

# Navigate to the application home page
driver.get("http://www.google.com")

# get the search textbox
search_field = driver.find_element_by_name('q')

search_field.clear()

# enter search keyword and submit
search_field.send_keys("Selenium WebDriver")
search_field.submit()


#-----------------------------------Pagination--------------------------------------

# Pagination
button = driver.find_element_by_xpath("//span[text()='Next']")
button.click() # page 1

button = driver.find_element_by_xpath("//span[text()='Next']")
button.click() # page 2

#-----------------------------------Browser level navigation-------------------------
driver.back()
driver.forward()
driver.refresh()
#-----------------------------------------------------------------------------------


# iterate through each element and print the text that is
# name of the search


# get the list of elements which are displayed after the search
# currently on result page using find_elements_by_class_name method
lists= driver.find_elements_by_class_name("LC20lb")

# get the number of elements found
print ("Found " + str(len(lists)) + " searches:")

i=0
for listitem in lists:
   print (listitem.get_attribute("innerHTML"))
   i=i+1
   if(i>10):
      break

# close the browser window
driver.quit()
```

## Take screenshots while scrolling and test headless (run selenium without showing the window)

```python
#------------------------Set-headless-------------------------------
from selenium.webdriver.firefox.options import Options


options = Options()
options.headless = True

driver = webdriver.Firefox(options=options)
#driver = webdriver.Firefox()
#--------------------------------------------------------------------


#geckodriver needed to be added to path in order for this to work
#driver = webdriver.Firefox()
driver.implicitly_wait(30)
driver.maximize_window()

# Navigate to the application home page
driver.get("http://www.google.com")

# get the search textbox
search_field = driver.find_element_by_name('q')

search_field.clear()

# enter search keyword and submit
search_field.send_keys("Selenium WebDriver")
search_field.submit()


#-----------------------------------Pagination--------------------------------------

# Pagination
button = driver.find_element_by_xpath("//span[text()='Next']")
button.click() # page 1

button = driver.find_element_by_xpath("//span[text()='Next']")
button.click() # page 2

#-----------------------------------Browser level navigation-------------------------
driver.back()
driver.forward()
driver.refresh()
#---------------------------------------Take a screenshot-----------------------------

body = driver.find_element_by_css_selector('body')
body.send_keys(Keys.PAGE_DOWN)
driver.save_screenshot("screenshot1.png")

body.send_keys(Keys.PAGE_DOWN)
driver.save_screenshot("screenshot2.png")

body.send_keys(Keys.PAGE_DOWN)
driver.save_screenshot("screenshot3.png")
#---------------------------------------------------------------------------------------



# get the list of elements which are displayed after the search
# currently on result page using find_elements_by_class_name method
lists= driver.find_elements_by_class_name("LC20lb")

# get the number of elements found
print ("Found " + str(len(lists)) + " searches:")

i=0
for listitem in lists:
   print (listitem.get_attribute("innerHTML"))
   i=i+1
   if(i>10):
      break

# close the browser window
driver.quit()


from IPython.display import Image

x=Image("./screenshot1.png")

y=Image("./screenshot2.png")

z=Image("./screenshot3.png")

display(x,y,z)
```

## Grabbing a cookie (from python scraping tutorial)

```python
from selenium import webdriver

driver = webdriver.Firefox(options=options)
driver.get("http://pythonscraping.com")
driver.implicitly_wait(10)
print(driver.get_cookies())

savedCookies = driver.get_cookies()

driver2 = webdriver.Firefox(options=options)
driver2.get("http://pythonscraping.com")
driver2.delete_all_cookies()
for cookie in savedCookies:
    driver2.add_cookie(cookie)

driver2.get("http://pythonscraping.com")
driver.implicitly_wait(10)
print(driver2.get_cookies())
```
